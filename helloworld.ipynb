```python
#Plotly Pandas Dataframe

import cufflinks as cf

pandas_df = cf.datagen.lines()

pandas_df.iplot() # More examples avilable at: https://plot.ly/ipython-notebooks/cufflinks/
```

```python
#Plotly Pandas Dataframe (select columns)

import pandas as pd
import numpy as np
import cufflinks as cf

df = pd.DataFrame(np.random.randn(1000, 2), columns=['A', 'B']).cumsum()

df.iplot(x='A', y='B') # More examples avilable at: https://plot.ly/ipython-notebooks/cufflinks/
```

```python
#Plotly Pandas Histogram

import pandas as pd
import cufflinks as cf

df = pd.DataFrame({'a': np.random.randn(1000) + 1,
                   'b': np.random.randn(1000),
                   'c': np.random.randn(1000) - 1})

df.iplot(kind='histogram') # More examples avilable at: https://plot.ly/ipython-notebooks/cufflinks/
```

```python
#Plotly Pandas BoxPlot

import pandas as pd
import cufflinks as cf

df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.iplot(kind='box') # More examples avilable at: https://plot.ly/ipython-notebooks/cufflinks/
```
